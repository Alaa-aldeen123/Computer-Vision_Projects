```python
img_path='/content/drive/MyDrive/AI course/Computer_Vesion/Open_CV/people.jpg'
img=cv2.imread(img_path)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
eye_classifier=cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_smile.xml'
    )
```

```python
eye=eye_classifier.detectMultiScale(
    gray_img,scaleFactor=1.1,minNeighbors=5,minSize=(20,20)
    )
eye
```

```python
import google.colab.patches
from google.colab.patches import cv2_imshow
```

```python
for x,y,w,h in eye:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

cv2_imshow(img)
```
