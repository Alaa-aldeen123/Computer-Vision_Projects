```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = 'dreamlike-art/dreamlike-photoreal-2.0'
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to('cpu')
```

```python
saved_images_folder='/content/drive/MyDrive/AI course/Computer_Vesion/Large_Vision_Models/Staible_Diffusion'
```

```python
prompts=['home in the forest,with breathtaking view',
        'villa in green hell, with wild life surround']
```

```python
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    image.save(f'{saved_images_folder}/picture_{i}.png')
```

```python
def generate_image(prompt):
    image = pipe(prompt).images[0]
    image.save(f'{saved_images_folder}/{prompt[:20]}.png')
```
